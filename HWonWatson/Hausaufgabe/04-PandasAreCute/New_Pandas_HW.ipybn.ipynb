```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
sdc=school_data_complete
sdc.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#GET THE FACTS
total_schools=len(pd.unique(sdc['school_name']))
print(f'The total number of schools is '+ str(total_schools))
total_students=len(pd.unique(sdc['Student ID']))
print(f'The total number of students in the district is ' + str(total_students))
total_budget=school_data['budget'].sum()
print(f'The total district budget is '+ str(total_budget))
avg_math=student_data['math_score'].mean()
print(f'The average math score is '+ str(avg_math))
avg_read=student_data['reading_score'].mean()
print(f'The average reading score is '+ str(avg_read))
passing_score=70.0
math_score=student_data['math_score']
percent_math=(((math_score)>= passing_score).sum()/(total_students/100))
print(f'The percentage of students passing math is %' + str(percent_math))
read_score=student_data['reading_score']
percent_read=(((read_score)>= passing_score).sum()/(total_students/100))
print(f'The percentage of students passing reading is %' + str(percent_read))
overall_pass=(percent_math + percent_read)/2
print(f'The overall percentage of students passing in the district is %'+ str(overall_pass))
```

```python
#Put into a Dataframe
district_df=pd.DataFrame({'Total Schools': [total_schools], 
                          'Total Students': [total_students], 
                          'Total Budget': [total_budget], 
                          'Average Math Score': [avg_math], 
                          'Average Reading Score': [avg_read], 
                          'Percent Passing Math': [percent_math],
                          'Percent Passing Reading': [percent_read],
                          'Overall Passing Rate': [overall_pass]
                          })
district_df.round(2)
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)
  
* Create a dataframe to hold the above results

```python
#Create Holding Cells for Data. Even if Data himself would never be trapped by one...
school_ids = pd.unique(sdc['School ID'])
schools_overall = {}
school_type = {}
type_index = ['type']
schools_index = [ 'budget', 'total_students',
        'budget_per_student', 'avg_math', 'avg_read', 'percent_math',
        'percent_read', 'pass_overall']
```

```python
#run the loop
for school in school_ids:
    _sd = sdc[sdc['School ID'] == school]
    _name = _sd["school_name"].iloc[0]
    _type = _sd["type"].iloc[0]
    _budget = _sd["budget"].iloc[0]
    _tot_students = len(pd.unique(_sd["Student ID"]))
    _budget_per_student = _budget / _tot_students
    _math = _sd["math_score"]
    _avg_math = _math.mean()
    _percent_math = (_math >= passing_score).sum() / (_tot_students / 100)
    _read = _sd["reading_score"]
    _avg_read = _read.mean()
    _percent_read = (_read >= passing_score).sum() / (_tot_students / 100)
    _pass_overall = (_percent_math + _percent_read) / 2
    schools_overall[_name]=[_budget, _tot_students, 
            _budget_per_student, _avg_math, _avg_read, _percent_math, 
            _percent_read, _pass_overall]
    school_type[_name]=[_type]
summary_df=pd.DataFrame(schools_overall, schools_index)
summary_df.round(2)
```

```python
#Clean it up
clean_summary=summary_df.transpose()
clean_summary.round(2)
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe

```python
#9th grade
columns=['school_name', 'grade', 'math_score']
stu_red=student_data.loc[:, columns]
ninth_group=stu_red.groupby('grade').get_group('9th')
ninth_school_group=ninth_group.groupby('school_name').mean()
ninth_table=ninth_school_group.rename(columns={'math_score': 'ninth_grade'})
ninth_table
```

```python
#10th grade
tenth_group=stu_red.groupby('grade').get_group('10th')
tenth_school_group=tenth_group.groupby('school_name').mean()
tenth_table=tenth_school_group.rename(columns={'math_score': 'tenth_grade'})
tenth_table
```

```python
#11th grade
elf_group=stu_red.groupby('grade').get_group('11th')
elf_school_group=elf_group.groupby('school_name').mean()
elf_table=elf_school_group.rename(columns={'math_score': 'eleventh_grade'})
elf_table
```

```python
#12th grade
zwolf_group=stu_red.groupby('grade').get_group('12th')
zwolf_school_group=zwolf_group.groupby('school_name').mean()
zwolf_table=zwolf_school_group.rename(columns={'math_score': 'twelfth_grade'})
zwolf_table
```

```python
nine_and_ten=pd.merge(ninth_table, tenth_table, how='left', on=['school_name'])
nine_elf=pd.merge(nine_and_ten, elf_table, how='left', on=['school_name'])
nine_zwolf=pd.merge(nine_elf, zwolf_table, how='left', on=['school_name'])
nine_zwolf.round(2)
```
